```python
def get_contract_creation(address):
    url = f"https://api.etherscan.io/api?module=contract&action=getcontractcreation&contractaddresses={address}&apikey={ETHERSCAN_API_KEY}"
    response = requests.get(url)
    
    try:
        data = response.json()
    except ValueError:
        print(f"[ERROR] Could not parse JSON for {address}")
        return {
            "Address": address,
            "BlockNumber": None,
            "TxHash": None
        }

    # DEBUG: Show the raw response if it's not valid
    if data.get("status") != "1" or not isinstance(data.get("result"), list):
        print(f"[ERROR] Etherscan returned an error for {address}")
        print("Full response:", data)
        return {
            "Address": address,
            "BlockNumber": None,
            "TxHash": None
        }

    try:
        result = data['result'][0]
        return {
            "Address": address,
            "BlockNumber": int(result['blockNumber']),
            "TxHash": result['txHash']
        }
    except (KeyError, IndexError, TypeError) as e:
        print(f"[WARNING] Couldn't parse result for {address}: {e}")
        print("Full result block:", data['result'])
        return {
            "Address": address,
            "BlockNumber": None,
            "TxHash": None
        }


# ‚è≥ Convert block number to timestamp
def get_block_timestamp(block_number):
    hex_block = hex(block_number)
    url = f"https://api.etherscan.io/api?module=proxy&action=eth_getBlockByNumber&tag={hex_block}&boolean=true&apikey={ETHERSCAN_API_KEY}"
    response = requests.get(url)    
    print(f"Block #{block_number} => URL: {url}")
    print("Block data:", response.json())

    try:
        timestamp_hex = response.json()['result']['timestamp']
        return int(timestamp_hex, 16)
    except:
        return None

# üîÅ Process each contract
results = []
for addr in contract_addresses:
    creation = get_contract_creation(addr)
    block_number = creation["BlockNumber"]

    if block_number is not None:
        ts = get_block_timestamp(block_number)
        if ts:
            dt = datetime.fromtimestamp(ts)
            age = (datetime.now() - dt).days
        else:
            dt, age = None, None
    else:
        dt, age = None, None

    results.append({
        "Address": addr,
        "BlockNumber": block_number,
        "CreationTimestamp": dt,
        "ContractAge": age
    })

    time.sleep(1)  # respect rate limit

# üíæ Save to CSV
df = pd.DataFrame(results)
df.to_csv("/Users/joshuasalter/Documents/GitHub/blockchain-data-portfolio/data/contract_creation_metadata.csv", index=False)
print("Saved contract creation data to 'data/contract_creation_metadata.csv'")
```

```python

df = pd.read_csv("/Users/joshuasalter/Documents/GitHub/blockchain-data-portfolio/data/contract_creation_metadata.csv")
```

```python
risky_opcodes = ['delegatecall', 'selfdestruct', 'callcode', 'tx.origin']
```

```python
# Flag opcodes in SourceCode
for opcode in risky_opcodes:
    col_name = f"Contains_{opcode}"
    df[col_name] = df['SourceCode'].str.contains(opcode, case=False, na=False)

# Total RiskScore based on presence of opcodes
df['RiskScore'] = df[[f'Contains_{op}' for op in risky_opcodes]].sum(axis=1)
```

```python
plt.figure(figsize=(8, 5))
df['RiskScore'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Distribution of Risk Scores")
plt.xlabel("Risk Score (number of risky opcodes)")
plt.ylabel("Number of Contracts")
plt.tight_layout()
plt.show()
```

```python
opcode_counts = df[[f'Contains_{op}' for op in risky_opcodes]].sum()

plt.figure(figsize=(8, 5))
opcode_counts.sort_values().plot(kind='barh', color='salmon')
plt.title("Frequency of Risky Opcodes in Verified Contracts")
plt.xlabel("Number of Contracts Using Opcode")
plt.tight_layout()
plt.show()
```

```python
# Optional: Fallback if 'ContractType' missing
if 'ContractType' not in df.columns:
    df['ContractType'] = "Unknown"

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='ContractType', y='RiskScore', palette='Set3')
plt.title("Risk Score by Contract Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='ContractAge', y='RiskScore', hue='ContractType', alpha=0.7)
plt.title("Risk Score vs Contract Age")
plt.xlabel("Contract Age (days)")
plt.ylabel("Risk Score")
plt.legend(title="Contract Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
